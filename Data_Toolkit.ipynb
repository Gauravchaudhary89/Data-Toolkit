#1 What is NumPy, and why is it widely used in Python?
ANS: NumPy is a Python library for numerical computing, offering fast, memory-efficient multidimensional arrays and tools for performing mathematical operations on them. It’s widely used because of its speed, vectorization, and interoperability with other data science libraries.

#2 How does broadcasting work in NumPy?
ANS: Broadcasting lets NumPy perform arithmetic operations on arrays of different shapes by automatically expanding their dimensions when possible, without copying data, making operations efficient and concise.

#3 What is a Pandas DataFrame?
ANS: A Pandas DataFrame is a 2-dimensional labeled data structure with columns of potentially different types, similar to a table in a database or Excel sheet.

#4 Explain the use of the groupby() method in Pandas.
ANS: The groupby() method splits data into groups based on a key, applies a function (like sum, mean), and then combines the results, useful for aggregation and analysis by categories.

#5 Why is Seaborn preferred for statistical visualizations?
ANS: Seaborn offers high-level, aesthetically pleasing plots with built-in statistical capabilities like regression lines, distribution plots, and heatmaps with minimal code.

#6 What are the differences between NumPy arrays and Python lists?
ANS: NumPy arrays are homogeneous and faster, supporting vectorized operations, while Python lists are heterogeneous, slower, and lack built-in mathematical operations.

#7 What is a heatmap, and when should it be used?
ANS: A heatmap is a color-coded matrix representation of data values. It's ideal for visualizing correlations, confusion matrices, or large datasets to spot patterns and outliers.

#8 What does the term “vectorized operation” mean in NumPy?
ANS: Vectorized operations allow batch processing of array elements without explicit loops, leading to faster, cleaner, and more efficient computations.

#9 How does Matplotlib differ from Plotly?
ANS: Matplotlib is a static, customizable plotting library ideal for simple to complex 2D plots. Plotly offers interactive, web-based plots with built-in tooltips, zoom, and export options.

#10 What is the significance of hierarchical indexing in Pandas?
ANS: Hierarchical indexing (MultiIndex) allows multiple index levels on rows or columns, enabling complex data structures and flexible data selection and reshaping.

#11 What is the role of Seaborn’s pairplot() function?
ANS: The pairplot() function creates a grid of scatter plots and histograms for visualizing relationships between multiple numerical variables at once.

#12 What is the purpose of the describe() function in Pandas?
ANS: The describe() function generates summary statistics (count, mean, std, min, quartiles, max) for numerical columns, giving a quick overview of data distribution.

#13 Why is handling missing data important in Pandas?
ANS: Unaddressed missing data can skew analyses and results. Pandas provides methods like fillna(), dropna(), and interpolate() to manage incomplete data effectively.

#14 What are the benefits of using Plotly for data visualization?
ANS: Plotly creates interactive, browser-based visualizations with zoom, hover, and export options, suitable for dashboards, web apps, and data presentations.

#15 How does NumPy handle multidimensional arrays?
ANS: NumPy supports n-dimensional arrays (ndarrays), enabling complex operations across multiple axes with efficient memory use and broadcasting.

#16 What is the role of Bokeh in data visualization?
ANS: Bokeh produces interactive, real-time web visualizations directly from Python, ideal for dashboards, streaming data, and linking multiple plots.

#17 Explain the difference between apply() and map() in Pandas.
ANS: map() works element-wise on Series, while apply() can apply a function to each row/column in a DataFrame or to each element in a Series.

#18 What are some advanced features of NumPy?
ANS: Advanced features include broadcasting, masked arrays, linear algebra operations, Fourier transforms, random sampling, and multidimensional indexing.

#19 How does Pandas simplify time series analysis?
ANS: Pandas offers specialized time-based indexing, resampling, date parsing, and moving window functions, making time series manipulation and analysis easy.

#20 What is the role of a pivot table in Pandas?
ANS: A pivot table reorganizes and summarizes data, enabling aggregation and comparison across multiple dimensions, similar to Excel’s pivot feature.

#21 Why is NumPy’s array slicing faster than Python’s list slicing?
ANS: NumPy arrays use contiguous memory blocks and fixed data types, enabling optimized, low-level operations, unlike Python lists which store references.

#22 What are some common use cases for Seaborn?
ANS: Seaborn is used for correlation heatmaps, categorical plots (box, violin, bar), distribution plots, pair plots, and regression analysis visualizations.

```python
#1 How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = np.sum(arr, axis=1)
print(row_sum)
```

```python
#2 Write a Pandas script to find the mean of a specific column in a DataFrame.

import pandas as pd

df = pd.DataFrame({'Marks': [80, 90, 70, 60]})
mean_value = df['Marks'].mean()
print(mean_value)
```

```python
#3 Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.show()
```

```python
#4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
#5 Generate a bar plot using Plotly.
import plotly.express as px

data = {'Category': ['A', 'B', 'C'], 'Values': [10, 20, 30]}
fig = px.bar(data, x='Category', y='Values')
fig.show()
```

```python
#6 Create a DataFrame and add a new column based on an existing column.
import pandas as pd

df = pd.DataFrame({'Marks': [50, 70, 90]})
df['Grade'] = df['Marks'].apply(lambda x: 'Pass' if x >= 60 else 'Fail')
print(df)
```

```python
#7 Write a program to perform element-wise multiplication of two NumPy arrays.
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print(result)
```

```python
#8 Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt

x = [1, 2, 3]
y1 = [2, 4, 6]
y2 = [1, 3, 5]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.show()
```

```python
#9 Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd

df = pd.DataFrame({'Marks': [50, 70, 85, 40]})
filtered_df = df[df['Marks'] > 60]
print(filtered_df)
```

```python
#10 Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 30, 40, 50]
sns.histplot(data, bins=5)
plt.show()
```

```python
#11 Perform matrix multiplication using NumPy.
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.dot(a, b)
print(result)
```

```python
#13 Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3],
    'y': [4, 5, 6],
    'z': [7, 8, 9]
})

fig = px.scatter_3d(df, x='x', y='y', z='z')
fig.show()
```
